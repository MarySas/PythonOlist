```python
# Create subset
sub = df[['product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'price', 'freight_value', 'payment_installments', 'payment_value', 'review_score', 'n_of_days', 'distance' ]]
```

```python
# Creating a correlation matrix using pandas
sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=10, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib. What it does is create an empty output

f,ax = plt.subplots(figsize=(8,8))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

```python
#The positive correlation are most related to price and products measurements and weight, price and freight_value and price and payment_value, 
#which can hint that the bigger the product, more the customer must pay for freight, making the total value higher.

# The weak (-0,25 to -0,5 ) negative correlation exists between review_scores and the number of days that the order was processed, shipped and delivered. 
# Is also present between payment_value and review_scores
```

```python
df.loc[df['review_score'] <= 2, 'review_category'] = 'Bad review'
```

```python
df.loc[df['review_score'] >= 4, 'review_category'] = 'Good review'
```

```python
df.loc[df['review_score'] == 3, 'review_category'] = 'Medium review'
```

```python
df['review_category'].value_counts(dropna = False)
```

```python
color_dict = dict({'Good review':'green',
                  'Medium review':'blue',
                  'Bad review': 'red'})
```

```python
# Create a categorical plot in seaborn using the review categories created above
g = sns.scatterplot(x="product_category_name", y="price", hue="review_category", data=df,
                  palette=color_dict, 
                   legend='full')
plt.xticks(color = 'black', rotation = 65, horizontalalignment = 'right')
sns.set(rc={"figure.figsize":(20, 5)}) #width=20, height=5
```
